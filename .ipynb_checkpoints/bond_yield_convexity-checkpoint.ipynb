```python
import pandas_datareader.data as web
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime.today()
```

```python
yields = web.DataReader(['DGS5', 'DGS10', 'DGS30', 'DAAA'], 'fred', start, end)
yields.columns = ['5Y Treasury', '10Y Treasury', '30Y Treasury', 'Corporate AAA']
yields.dropna(how='all', inplace=True)
```

```python
yields.plot(figsize=(12, 6), title='Bond Yields Over Time')
plt.ylabel("Yield (%)")
plt.grid(True)
plt.show()
```

```python
snapshot = yields.dropna().iloc[-1]
print("Most Recent Yields:")
print(snapshot)
```

```python
def bond_price(face, coupon_rate, years, ytm, freq=2):
    periods = int(years * freq)
    coupon = coupon_rate * face / freq
    discount_factors = [(1 + ytm / freq) ** -t for t in range(1, periods + 1)]
    pv_coupons = sum([coupon * df for df in discount_factors])
    pv_face = face / (1 + ytm / freq) ** periods
    return pv_coupons + pv_face

def macaulay_duration(face, coupon_rate, years, ytm, freq=2):
    periods = int(years * freq)
    coupon = coupon_rate * face / freq
    cashflows = [coupon] * (periods - 1) + [coupon + face]
    times = [t for t in range(1, periods + 1)]
    discount_factors = [(1 + ytm / freq) ** -t for t in times]
    weighted_times = [t * cf * df for t, cf, df in zip(times, cashflows, discount_factors)]
    bond_val = bond_price(face, coupon_rate, years, ytm, freq)
    return sum(weighted_times) / (bond_val * freq)

def convexity(face, coupon_rate, years, ytm, freq=2):
    periods = int(years * freq)
    coupon = coupon_rate * face / freq
    cashflows = [coupon] * (periods - 1) + [coupon + face]
    times = [t for t in range(1, periods + 1)]
    discount_factors = [(1 + ytm / freq) ** -t for t in times]
    weighted_times_sq = [t * (t + 1) * cf * df for t, cf, df in zip(times, cashflows, discount_factors)]
    bond_val = bond_price(face, coupon_rate, years, ytm, freq)
    return sum(weighted_times_sq) / (bond_val * (freq ** 2) * (1 + ytm / freq) ** 2)
```

```python
face = 1000
coupon_rate = 0.05
years = 10
ytm = snapshot['10Y Treasury'] / 100  # Convert % to decimal

price = bond_price(face, coupon_rate, years, ytm)
duration = macaulay_duration(face, coupon_rate, years, ytm)
conv = convexity(face, coupon_rate, years, ytm)

print(f"Price: ${price:.2f}")
print(f"Duration: {duration:.2f} years")
print(f"Convexity: {conv:.2f}")
```

```python
import numpy as np

ytms = np.linspace(ytm - 0.01, ytm + 0.01, 100)
prices = [bond_price(face, coupon_rate, years, y) for y in ytms]

plt.figure(figsize=(10, 5))
plt.plot(ytms * 100, prices)
plt.title("Bond Price vs Yield (Convexity Visualized)")
plt.xlabel("Yield to Maturity (%)")
plt.ylabel("Bond Price ($)")
plt.axvline(ytm * 100, color='gray', linestyle='--', label='Current YTM')
plt.grid(True)
plt.legend()
plt.show()
```

```python
bonds = [
    {"name": "Short-Term Low Coupon", "coupon": 0.02, "years": 3},
    {"name": "Long-Term Low Coupon", "coupon": 0.02, "years": 30},
    {"name": "Short-Term High Coupon", "coupon": 0.07, "years": 3},
    {"name": "Long-Term High Coupon", "coupon": 0.07, "years": 30},
    {"name": "Par Bond", "coupon": 0.05, "years": 10},
]

ytm = snapshot['10Y Treasury'] / 100  # current market yield
face = 1000

results = []

for bond in bonds:
    price = bond_price(face, bond["coupon"], bond["years"], ytm)
    dur = macaulay_duration(face, bond["coupon"], bond["years"], ytm)
    conv = convexity(face, bond["coupon"], bond["years"], ytm)
    results.append({
        "Name": bond["name"],
        "Price": round(price, 2),
        "Duration": round(dur, 2),
        "Convexity": round(conv, 2)
    })

df = pd.DataFrame(results)
df
```

```python
def price_change_approx(price, duration, convexity, delta_y):
    return -duration * delta_y + 0.5 * convexity * (delta_y ** 2)

# Sample bond
coupon = 0.05
years = 10
price = bond_price(face, coupon, years, ytm)
dur = macaulay_duration(face, coupon, years, ytm)
conv = convexity(face, coupon, years, ytm)

# Test +/- 100bps
for delta in [-0.01, 0.01]:
    exact = bond_price(face, coupon, years, ytm + delta)
    approx = price + price * price_change_approx(price, dur, conv, delta)
    print(f"ΔYTM: {delta*100:+.0f}bps | Exact: {exact:.2f}, Approximated: {approx:.2f}")
```

```python
ytms = np.linspace(ytm - 0.01, ytm + 0.01, 100)
plt.figure(figsize=(12, 6))

for bond in bonds:
    prices = [bond_price(face, bond["coupon"], bond["years"], y) for y in ytms]
    plt.plot(ytms * 100, prices, label=bond["name"])

plt.title("Bond Price vs Yield")
plt.xlabel("Yield to Maturity (%)")
plt.ylabel("Price ($)")
plt.axvline(x=ytm * 100, color='gray', linestyle='--', label='Current YTM')
plt.legend()
plt.grid(True)
plt.show()
```
